```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

** Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. Remeber back to your matplotlib lecture...**

```python
df3.plot.scatter(x='a',y='b',c='red',s=50,figsize=(12,3))
```

** Create a histogram of the 'a' column.**

```python
df3['a'].plot.hist(alpha=0.4, edgecolor = 'black',  linewidth=1)
```

** These plots are okay, but they don't look very polished. I will use 'ggplot' and redo the histogram from above.***

```python
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(alpha=0.5,bins=25, edgecolor = 'white',  linewidth=1)
```

## Boxplot

```python
df3[['a','b']].plot.box()
```

## kde plot

```python
df3['d'].plot.kde()
```

## Density plot

```python
df3['d'].plot.density(lw=5,ls='--')
```

## Area plot

```python
df3.iloc[0:30].plot.area(alpha=0.4)
```

** I have noticed that the legend wasn't in a very stylish position, so I've decided to put it out of the plot. **

[Full Documentation Reference](https://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas)

```python
f = plt.figure()
df3.ix[0:30].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
